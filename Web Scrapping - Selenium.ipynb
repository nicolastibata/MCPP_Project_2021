### Usaquén

#### Getting the Driver 

```python
driver_usaquen = webdriver.Chrome('/Users/Nicolas/Downloads/chromedriver')
driver_usaquen.get('https://www.olx.com.co/usaquen_g4300120/apartamentos-casas-venta_c367') # Automated text 
```

---------------

#### Getting the Button

```python
boton = driver_usaquen.find_element_by_xpath('//button[@data-aut-id="btnLoadMore"]') # Para conserguir el botón de cargar más
```

In this case, we need too `create a button` to get more data, BUT is important to click the button in a random way nor you get blocked

```python
# Se quiere obtener la mayor cantidad de obsrevaciones, por lo tanto es necesario hacer más de 10 clicks
for i in range(50): #500 para un total de 10000 observaciones
    try:
        boton.click()
        # Se utiliza sleep de manera random para humanizar la máquina y no detectar que es un bot
        sleep(random.uniform(5.0,20.0))
        boton = driver_usaquen.find_element_by_xpath('//button[@data-aut-id="btnLoadMore"]')
    except: 
        break
```

-----------------------

```python
links_usaquen = driver_usaquen.find_elements_by_xpath('//*[@href]')
```

#### Getting the links of the properties

```python
data_links_usaquen = []
for link in links_usaquen:
    data = link.get_attribute('href')
    data_links_usaquen.append(data)
    
data_links_usaquen
```

```python
len(data_links_usaquen)
```

```python
driver_usaquen.quit() # Even faster even better
```

```python
base_links = []
for data in data_links_usaquen:
    if data.find("iid") != -1:
        base_links.append(data)
    else:
        continue
```

```python
base_links
```

```python
len(base_links)
```
