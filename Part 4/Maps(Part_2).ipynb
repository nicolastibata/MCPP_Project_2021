```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('df_maps_final.csv')
```

```python

gf2 = gf.rename(columns={'Localidad':'localidad'})
gf2
```

```python
bogotamap2 = gpd.read_file('poligonos-localidad-2.shp')
bogotamap2
```

```python
nombre = bogotamap2.columns.values
nombre
```

```python
bogotamap3 = bogotamap2.rename(columns={'nombre_de_l':'localidad','area_de_la_':'area','acto_admini':'acto'})
bogotamap3
```

```python
bogotamap4 = bogotamap3.drop(['acto','area','identificad'],axis=1)
bogotamap4
```

```python
mapa_estadisticas2 = bogotamap4.merge(gf2, on = 'localidad')
mapa_estadisticas2
```

```python
fig, ax = plt.subplots(1, figsize=(25,25))
plt.xticks(rotation=90)
mapa_estadisticas2.plot(column='Ingreso promedio mensual por persona (COP)*', cmap="Blues", linewidth=0.5, ax=ax, edgecolor ='.4')
ax.axis('off') #que no se muestren los ejes
ax.set_title ('Ingreso promedio mensual por persona según la localidad',
              fontdict={'fontsize': '25', 'fontweight':'2'})
plt.savefig('Ingreso_por_localidades')
```
