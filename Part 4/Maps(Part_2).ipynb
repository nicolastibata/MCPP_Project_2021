```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('df_maps_final.csv')
```

```python
df
```

```python
df = df.rename(columns={'Localidad':'localidad'})
df
```

```python
bogotamap2 = gpd.read_file('poligonos-localidad-2.shp')
bogotamap2
```

```python
nombre = bogotamap2.columns.values
nombre
```

```python
bogotamap3 = bogotamap2.rename(columns={'nombre_de_l':'localidad','area_de_la_':'area','acto_admini':'acto'})
bogotamap3
```

```python
bogotamap4 = bogotamap3.drop(['acto','area','identificad'],axis=1)
bogotamap4
```

```python
mapa_estadisticas2 = bogotamap4.merge(df, on = 'localidad')
mapa_estadisticas2
```

```python
mapa_estadisticas2 = mapa_estadisticas2.drop(['Unnamed: 0'], axis = 1)
```

```python
mapa_estadisticas2
```

```python
fig, ax = plt.subplots(1, figsize=(25,25))
plt.xticks(rotation=90)
mapa_estadisticas2.plot(column='Promedio', cmap="Blues", linewidth=0.5, ax=ax, edgecolor ='.4', legend = True)
ax.axis('off') #que no se muestren los ejes
mapa_estadisticas2.apply(lambda x: ax.annotate(text=x['localidad'], xy=x.geometry.centroid.coords[0], ha='center'), axis=1);
ax.set_title ('Precio promedio de propiedades por localidad',
              fontdict={'fontsize': '25', 'fontweight':'2'})
#plt.savefig('Ingreso_por_localidades')
```

```python
fig, ax = plt.subplots(1, figsize=(25,25))
plt.xticks(rotation=90)
mapa_estadisticas2.plot(column='Homicidios', cmap="Reds", linewidth=0.5, ax=ax, edgecolor ='.4', legend = True)
ax.axis('off') #que no se muestren los ejes
mapa_estadisticas2.apply(lambda x: ax.annotate(text=x['localidad'], xy=x.geometry.centroid.coords[0], ha='center'), axis=1);
ax.set_title ('Número de homicidios por localidad',
              fontdict={'fontsize': '25', 'fontweight':'2'})
#plt.savefig('Ingreso_por_localidades')
```
